```python
#local audio file request 1 - automated(list)
url_list = []

import requests
def read_file(filename, chunk_size=5242880):
    with open(filename, 'rb') as _file:
        while True:
            data = _file.read(chunk_size)
            if not data:
                break
            yield data

for i in range(len(audio_file)):
  filename = audio_file[i]
  json = {"speaker_labels": True}
  headers = {'authorization': "f330e85d644243cd8c2b8506b312c55d"}
  response = requests.post('https://api.assemblyai.com/v2/upload', json = json, headers=headers, data=read_file(filename))
  url_list.append(response.json()['upload_url'])
  print(response.json()['upload_url'])

print(url_list)
```

```python
#local audio file request 2 - automated(list)

id_list = []

for i in range(len(audio_file)):
  endpoint = "https://api.assemblyai.com/v2/transcript"
  json = { "audio_url": url_list[i] , "speaker_labels" : True}
  headers = {
      "authorization": "f330e85d644243cd8c2b8506b312c55d",
      "content-type": "application/json"
  }
  response = requests.post(endpoint, json=json, headers=headers)
  print(response.json())
  print(response.json()['id'])
  id_list.append(response.json()['id'])
```

```python
endpoint_test = "https://api.assemblyai.com/v2/transcript/" + id_list[0]
print(endpoint_test)
```

```python
#local audio file return - automated(list)
final = []
for i in range(len(audio_file)):
  endpoint = "https://api.assemblyai.com/v2/transcript/" + id_list[i]
  headers = {
      "authorization": "f330e85d644243cd8c2b8506b312c55d",
  }
  response = requests.get(endpoint, headers=headers)
  print(response.json())
  final.append(response.json()['words'])

#print(final)
```

```python
print(final[0])
print(final[0][0])
print(final[0][0]['text'])
```

```python
num_speakers=[]

for i in range(len(final)):
  num_speakers.append([])

for i in range(len(final)):
  for j in range(len(final[i])):
      if ([final[i][j]['speaker']] not in num_speakers[i]):
          num_speakers[i].append([final[i][j]['speaker']])
        
print(num_speakers)
```

```python
ns = []

for i in range(len(final)):
  ns.append([])

for i in range(len(final)):
  for j in range(len(num_speakers[i])):
      ns[i].append([j])
ns
```

```python
for i in range(len(final)):
  for j in range(len(final[i])):
      a = num_speakers[i].index([final[i][j]['speaker']])
      #num_speakers[a].append(text_json['words'][i]['text'])
      ns[i][a] = ns[i][a] + [final[i][j]['text']]

ns
```

```python
for i in range(len(final)):
  for j in range(len(ns[i])):
    ns[i][j].remove(j)
ns
```
